```python
import turtle
import time
from matplotlib import colors
t = turtle.Turtle()
t.speed(0)
t.penup()

flag_height = 250 
flag_width = 475

stripe_height = flag_height/13  #Establishes the required height for the stripes
stripe_width = flag_width      #The stripe length is the same as the length of the flag

start_x = -237
start_y = 125
y = start_y - stripe_height #This is for later in the program so that the stripes can have space between them
star_size = 10
#draw_renctangle(length,height,color)
def draw_rectangle(length,height, color):
    
    t.pendown()    #Starts the drawing
    t.color(color)   #Sets the color
    t.begin_fill()   #Fills the rectangle with the color
    t.forward(length) #Begins tracing to the desired width
    t.right(90)     #Turns the pen so that it can begin working downward
    t.forward(height) #Traces do to the desired height
    t.right(90)
    t.forward(length)
    t.right(90)
    t.forward(height)
    t.right(90)
    t.end_fill()
    t.penup()

square_height = (7/13)* flag_height #Specified measurments for the blue square
square_width = (0.76)* flag_height

def draw_square(x, y, height, width, color):
    t.goto(x,y)
    t.pendown()
    t.color(color)
    t.begin_fill()
    t.forward(width)
    t.right(90)
    t.forward(height)
    t.right(90)
    t.forward(width)
    t.right(90)
    t.forward(height)
    t.right(90)
    t.end_fill()
    t.penup()

    
    #draw_flag(height)
def draw_flag(height):  #Outlines the flag
    t.pendown()
    t.color('black')  
    t.forward(height*1.9)
    t.right(90)
    t.forward(height)
    t.right(90)
    t.forward(height*1.9)
    t.right(90)
    t.forward(height)
    t.right(90)
    t.penup()

def draw_star(color,size) :   #draws the stars
    t.color(color)
    t.setheading(0)
    t.pendown()
    for turn in range(0,5):
        t.forward(size)
        t.right(144)
        t.forward(size)
        t.right(144) 
    t.penup()

def draw_six_stars_rows():    #draws rows of six stars
    gap_between_stars = 30
    gap_between_lines = stripe_height + 6
    y = 112
    for row in range(0,5) :
        x = -222
        for star in range (0,6) :
            t.goto(x,y)
            draw_star('white', star_size)
            
            x = x + gap_between_stars
        y = y - gap_between_lines


def draw_five_stars_rows():    #Draws rows of five stars
    gap_between_stars = 30
    gap_between_lines = stripe_height + 6
    y = 100
    for row in range(0,4) :
        x = -206
        for star in range (0,5) :
            t.goto(x,y)
            draw_star('white', star_size)
            x = x + gap_between_stars
        y = y - gap_between_lines

def get_color(color):
    
    print(colors.to_rgb(color))
    
        
def main():     #Calls all the functions needed to make flag
    y = start_y
    t.goto(start_x,start_y)
    draw_flag(flag_height)
    for stripe in range(0,7):   #draws 7 stripes             
        draw_rectangle(stripe_width,stripe_height,'red')
        y = y - stripe_height*2
        t.goto(start_x, y)
    draw_square(start_x,start_y,square_height,square_width,'navy')
    draw_five_stars_rows()
    draw_six_stars_rows()
    get_color('red')

    
main()
```
